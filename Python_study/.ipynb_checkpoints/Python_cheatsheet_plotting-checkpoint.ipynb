# Histograms

> 도수분포표 vs 히스토그램
- 도수분포표 : 
    - 특정 구간에 속하는 자료의 개수를 표현 
    - 히스토그램, 라인 차트, 막대 그래프, 원그래프로 표현
- 히스토그램 : 
    - 도수분포표를 시각화하여 막대그래프로 표현, 하지만 bar plot과는 다름 

> 막대그래프 vs 히스토그램 (bar plot vs hist plot)
- bar plot : 
    - 표현 값에 비례하여 높이와 길이를 지닌 직사각형 막대로 범주형 데이터를 표현하는 차트나 그래프, 합계, 평균 등의 수치를 시각화 
    - 계급(즉 가로)을 생각하지 않고 세로의 높이로만 나타냄
- hist plot : 
    - 구간별 빈도수를 표현
    - 가로와 세로를 함께 생각해야함

> 히스토그램 vs 정규분포 (hist plot vs density plot)
- hist plot : 구간별 빈도수
- density plot : 확률 밀도 함수 - 확률 변수의 밀도를 표현

```python
df3 = pd.DataFrame({'a' : np.random.randn(1000)+1,
                    'b' : np.random.randn(1000),
                    'c' : np.random.randn(1000)-1},
                   columns = ['a','b','c'])
df3
```

## alpha
- 투명도

```python
df3.plot.hist(alpha = 0.5)
plt.show()
```

## stacked & bins
- stacked : 
    - 위에 쌓아서 그림
- bins : 
    - box의 수 (가로를 자르는 갯수)
    - default : 10

```python
df3.plot.hist(stacked = True, bins = 20)
plt.show()
```

## orientation
- orientation : 
    - orientation = 'horizontal' :  축을 수평으로
- cumulative : 
    - 누적

```python
df3['a'].plot.hist(orientation = 'horizontal', cumulative = True)
```

## diff
- diff : 차분 (전 값과의 차이)

```python
df3['a_diff'] = df3['a'].diff()
df3['a_shift'] = df3['a'].shift(1)
df3['a_minus'] = df3['a'] - df3['a_shift']
df3[['a_diff', 'a_minus']]
```

```python
df3['a'].diff().hist()
```

```python
df3 = df3[['a','b','c']]
```

```python
df3.diff().hist(color='k', alpha=0.5, bins = 30)
plt.show()
```

```python
dat = pd.DataFrame({'a': np.random.randn(1000),
                    'b' : np.random.randint(0,4,1000)})
dat.head()
```

```python
dat['a'].hist(by = dat['b'], figsize = (6,4))
plt.show()
```

# Hexbin plot
> Hexagonal Bin Plot
- 데이터가 클 때 각각의 점을 산점도로 표현할 때의 단점을 보완
- 값의 많고 적음을 비교할 수 있음 (색으로 판단)
- Histogram이랑 scatter plot 합친거라고 생각

```python
df = pd.DataFrame(np.random.randn(1000,2),
                 columns = ['a', 'b'])
df['b'] = df['b'] + np.arange(1000)
df
```

```python
df.plot.hexbin(x = 'a', y = 'b', gridsize = 25)
```

```python
df['z'] = np.random.uniform(0, 3, 1000)
df
```

```python
df.plot.hexbin(x = 'a', y = 'b', C = 'z',
              reduce_C_function=np.max,
              gridsize = 25)
```

# Scatter Matrix plot
> alpha : 
- 투명도

> diagonal = 'kde' :
- kde : kernel density estimation (커널 밀도 추정)

```python
from pandas.plotting import scatter_matrix

df = pd.DataFrame(np.random.randn(1000,4),
                 columns = ['a', 'b', 'c','d'])
df.head()
```

```python
scatter_matrix(df, 
               alpha = 0.2,
               figsize = (6,6),
               diagonal = 'kde')
plt.show()
```

# Kernel Density Estimate plot
> Kernel density estimate
- 확률밀도함수
- 이산형 자료를 정규분포 형태로 만들어 줌

> - 통계에서 커널 밀도 추정 (KDE)은 임의 변수의 확률 밀도 함수 (PDF)를 추정하는 비모수적 방법이다. 
- 이 함수는 가우스 커널을 사용하며 자동 대역폭 결정을 포함한다.
- 커널 밀도 추정치는 히스토그램 과 밀접한 관련이 있지만 적절한 커널을 사용하여 매끄럽고 연속성과 같은 속성을 부여 할 수 있다.
- 커널 함수는 원점을 중심으로 대칭이며 적분값이 1인 함수다.

## density plot
> density plot
- 이산형 자료를 끊이지 않고 부드럽게 표현하는 plot
- Histogram -> 곡선으로 표현
- Histogram의 y축 -> frequency
- Density의 y축 -> Density

- plot.density(), plot.kde() : 동일한 plot

```python
ser = pd.Series(np.random.randn(1000))
ser
```

```python
ser.plot()
```

```python
ser.plot.hist()
```

```python
ser.plot.density()
```

```python
ser.plot.kde()
```
