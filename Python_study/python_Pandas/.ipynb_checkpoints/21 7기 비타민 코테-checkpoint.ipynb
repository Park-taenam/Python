```python
#문제 1
중간고사 = [34, 30, 26, 40, 33, 15, 31, 21, 17, 40]
기말고사 = [45, 48, 25, 50, 50, 28, 39, 33, 47, 42]
과제 = [5, 10, 8, 10, 10, 7, 7, 9, 10, 2]

a = list()

for x, y, z in zip(중간고사, 기말고사, 과제):
    a.append(x+y+z)
    
a.sort(reverse=True)
a
```

```python
#문제2
sn = input('지원번호를 입력하세요: ')

if (sn[3:4] == "1" or sn[3:4] == "3"):
    print("성별: 남")
elif (sn[3:4] == "2" or sn[3:4] == "4"):
    print("성별: 여")

year = sn[0:2]
semester = sn[2:3]

if (sn[3:4] == "1" or sn[3:4] == "2"):
    print("기수: " + str((int(year) - 18) * 2 + int(semester)))
    print("편입: x")
elif (sn[3:4] == "3" or sn[3:4] == "4"):
    print("기수: " + str((int(year) - 18) * 2))
    print("편입: o")

if (sn[-1] == "s"):
    print("운영진: o")
else:
    print("운영진: x")
```
