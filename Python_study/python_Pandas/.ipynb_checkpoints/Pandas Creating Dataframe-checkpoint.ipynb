```python
df = pd.DataFrame(
        [[4 ,5, 6],
         [7, 8, 9],
         [10, 11, 12]],
        index = [1, 2, 3],
        columns=['a','b','c']) #리스트 형태의 데이터를 series라고 부름
df
```

```python
print(df["c"],'\n') #대소문자 생각

print(df[["a","c"]],'\n')

print(df.loc[1]) #행
```

```python
df.loc[[1,2],["a","c"]] #행, 열순서
```

```python
#두 개의 인덱스 (tuole로 지정.)
#MultiIndex
df = pd.DataFrame(
    {"a" : [4 ,5, 6],
    "b" : [7, 8, 9],
    "c" : [10, 11, 12]},
    index = pd.MultiIndex.from_tuples(
    [('d',1),('d',2),('e',2)],
    names=['n','v']))
df
```

```python
print(df[df.b >7],'\n') #df.b >7 boolean 형태로 나옴.
                        #df['b']>7 와 동일
    
print(df.b > 7,'\n')

print(df['b'] > 7)
```

```python
df = pd.DataFrame(
    {"a" : [4 ,5, 6, 6],
    "b" : [7, 8, 9, 9],
    "c" : [10, 11, 12, 12]},
    index = pd.MultiIndex.from_tuples(
    [('d',1),('d',2),('e',2),('e',3)],
    names=['n','v']))

df.drop_duplicates() #중복된 행을 제거해준다.
#df.drop_duplicates(inplace=True) 원래의 dataframe을 변경시킨다. 
#df = df.drop_duplicates() 보통 이렇게 사용
#df.drop_duplicates? #어떤 parameter를 사용할 수 있는지 나온다.
```

```python
#isin : 특정 값 유무
df.a.isin([5])  #리스트 형태로 작성해야함
                #df["a"].isin([5]) 동일
```

```python
import numpy as np
df = pd.DataFrame(
    {"a" : [4 ,5, 6, 6, np.nan],
    "b" : [7, 8,np.nan, 9, 9],
    "c" : [10, 11, 12, np.nan,12]},
    index = pd.MultiIndex.from_tuples(
    [('d',1),('d',2),('e',2),('e',3),('e',4)],
    names=['n','v']))
df
```

```python
pd.isnull(df)  #df.isnull() 도 동일
#df['a'].isnull().sum()  : null의 갯수
```

```python
pd.notnull(df) #df.notnull() 동일
#df['a'].isnull().sum()  : notnull의 갯수
```
