```python
#ifelse
#n1
score=int(input("성적을 입력하시오: "))
if score >= 60:
    print("합격입니다")
else:
    print("불합격입니다")
    
#n2
num = int(input("정수를 입력하시오: "))
if num%2==0:
    print("짝수입니다")
else:
    print("홀수입니다")
    
#n3
score = int(input("점수를 입력하시오:"))
if score >= 90 :
    print("합격입니다.")
    print("장학금도 받을 수 있습니다.")
    
#n4
age = int(input("나이를 입력하시오: "))
if age >= 15:
    print("이 영화를 보실 수 있습니다.")
else:
    print("이 영화를 보실 수 없습니다.")

#n5
year = int(input("연도를 입력하시오: "))
if((year % 4 ==0 and year % 100 != 0) or year % 400 == 0):
    print(year, "년은 윤년입니다.")
else:
    print(year, "년은 윤년이 아닙니다.")


#n6
import random

    #random.randrange(6) 0에서 5까지 정수를 랜덤하게 생성할 수 있다.
    
print("동전 던지기 게임을 시작합니다.")
coin = random.randrange(2)
if coin == 0:
    print("앞면입니다.")
else :
    print("뒷면입니다.")
print("게임이 종료되었습니다.")

#n7
num=int(input("정수를 입력하시오: "))
if num > 0:
    print("양수입니다.")
elif num == 0:
    print("0입니다.")
else:
    print("음수입니다.")
    
#n8
import random
time = random.randint(1,24)             #random.randint(1,24) 1,24 사이의 정수 랜덤추출.
print("좋은 아침입니다. 지금 시작은" + str(time) + "시 입니다.")
sunny = random.choice([True, False])   #random.choice([,])   둘 중 하나 랜덤으로 선택.
if sunny:
    print("현재 날씨가 화창합니다. ")
else:
    print("현재 날씨가 화창하지 않습니다. ")
    
if time >= 6 and time<9 and sunny:
    print("종달새가 노래를 한다.")
else:

    print("종달새가 노래를 하지 않는다.")
    
#n9
num = int(input("정수를 입력하시오: "))
if num >= 0:
    if num==0:
        print("0입니다")
    else:
        print("양수입니다.")
else:
    print("음수입니다.")
    
#n10   (**)
options=["왼쪽","중앙","오른쪽"]
computer_choice = random.choice(options)
user_choice=input("어디를 수비하시겠어요?(왼쪽, 중앙, 오른쪽)")
if computer_choice == user_choice:
    print("수비에 성공하셨습니다.")
else:
    print("페널트 킥이 성공하였습니다.")
```
