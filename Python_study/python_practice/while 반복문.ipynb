```python
import random

tries = 0
guess = 0;
answer = random.randint(1,100)

print("1부터 100 사이의 숫자를 맞추시오.")

while guess != answer:
    guess = int(input("숫자를 입력하시오: "))
    tries = tries+1
    if guess < answer:
        print("낮음!")
    elif guess > answer:
        print("높음!")
        
if guess == answer:
    print("축하합니다. 시도횟수 =",tries)
    print("정답은 ",answer)
    
################################################################    
    
breads = ["호밀빵", "위트", "화이트"]
meats = ["미트볼", "소시지", "닭가슴살"]
vegis = ["양상추", "토마토","오이"]
sauces = ["마요네즈", "허니 머스타드", "칠리"]

for b in breads:
    for m in meats:
        for v in vegis:
            for s in sauces:
                print(b+"+"+m+"+"+v+"+"+s)     #모든 조합이 다 나온다.
                
```
