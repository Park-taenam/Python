```python
###과제1

import random

print("부루마블 게임\n")

count = -1
count1 = 0
num = 0

for i in range(1,3):
    print("엔터키를 누르면 주사위를 던집니다.\n")
    a=random.randint(1,6)
    b=random.randint(1,6)
    count = count + 2
    count1 = count1 + 2
    if a==b:
        print("%d 번째 주사위 : %d"%(count, a))
        print("%d 번째 주사위 : %d\n"%(count1, b))
        num = num+a+b
        print("같은 수가 나왔습니다.")
    else:
        print("%d 번째 주사위 : %d"%(count, a))
        print("%d 번째 주사위 : %d"%(count1, b))
        num = num+a+b
        break
print("")
print("%d칸 이동하세요!!" %num)
```
