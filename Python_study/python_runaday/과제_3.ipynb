```python
##과제_1
##5개만 실행

print("영어 퀴즈 프로그램입니다.")

score = 0
b = 0
dict = {"사과" : "apple", "코끼리" : "elephant", "별": "star" , "책상" : "desk", "연필" : "pencil"}
question = ["사과", "코끼리", "별", '책상', '연필']
ans = ["apple", "elephant", "star", "desk", "pencil"]

def quiz(word):
    global score
    global b
    global dict
    a=input(str(b) + "번 문제) " + word + " : ")
    if a == dict[word]:
        print("정답입니다.")
        score= score + 10
        print("현재 점수 : " + str(score) + "\n")
    else:
        print("오답입니다. 다시 풀어주세요.")
        score = score - 5
        print("현재 점수 : "  + str(score) + "\n")
        quiz(word)
        

for i in range(5):
    b = i+ 1
    quiz(question[i])
    
print("총 획득한 점수는 " + str(score) + "점 입니다.")
```

```python
##과제_4

from statistics import *

print("통계 프로그램")
print(" 1. 평균 \n 2. 최댓값 \n 3. 최솟값 \n 4. 분산 \n 5. 표준편차 \n 6. 사용설명보기 \n 9. 프로그램 종료")




while True:
    a = list(map(int,input("값을 입력하세요 : (예시_45 23 63 75 23) >>").split())) #리스트에 int형으로 입력
    b = int(input("알고 싶은 통계를 선택하세요. >> "))
    
    if b == 1:
        result = mean(a)
        print("평균은 " + str(result) + "입니다.")

    elif b==2:
        result = max(a)
        print("최댓값은 " + str(result) + "입니다.")

    elif b==3:
        result = min (a)
        print("최솟값은 " + str(result) + "입니다.")

    elif b==4:
        result = variance(a)
        print("분산은 " + str(result) + "입니다.")

    elif b==5:
        result = stdev(a)
        print("표준편차는 " + str(result) + "입니다.")

    elif b==6:
        print(" 1. 평균 \n 2. 최댓값 \n 3. 최솟값 \n 4. 분산 \n 5. 표준편차 \n 6. 사용설명보기 \n 9. 프로그램 종료")

    else:
        print("프로그램 종료합니다.")
        break
```

```python
##과제_5

class Book:
    ## 생성자
    def __init__(self, name, author, publisher):
        # 책이름
        self.name=name
        # 작가
        self.author=author
        # 출판사
        self.publisher=publisher
        # 대여 횟수
        self.number=0
        # 대여 가능 여부
        self.rentable=True
        # 대여자
        self.myid=""
        
    ## 대출 메서드
    def rent(self, day, myid):
        if self.rentable==True:
            print("대출날짜: {} " .format(day))
            print("대여자: {} " .format(myid))
            self.number= self.number +1
            self.rentable=False
            self.myid=myid
        else:
            print("현재 대출이 불가합니다")
            
    ## 반납 메서드
    def turnin(self, myid):
        if (self.rentable==False) and (self.myid==myid):
            print("반납이 완료되었습니다")
            self.rentable=True
            self.myid=""
        else:
            print("반납이 불가합니다. 안내데스크에 문의바랍니다")
            
    ## 확인 메서드
    def identify(self):
        print("현재 대여자 : {}" .format(self.myid))
        print("대여횟수  : {}" .format(self.number))

ex = Book("harry", "dontknow", "dontknow2")
print(ex.rent(20200921, "박태남"))
print(ex.turnin("박태남"))
print(ex.rent(20200921, "박태남"))
print(ex.turnin("박태남"))
print(ex.number)
print(ex.identify())
```
